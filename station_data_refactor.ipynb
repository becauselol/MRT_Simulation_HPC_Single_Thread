```python
with open('data/input/station_data.csv', 'r') as file:
    data = file.read()
```

```python
rows = data.split("\n")
```

```python
split_more = [i.split(",") for i in rows]
```

```python
target = [i[0] for i in split_more]
```

```python
target
```

```python
new_result = []
for i in target:
    row = i.split(" ")
    temp = row[-1]
    rest = " ".join(row[:-1])
    rest = rest + "," + temp
    new_result.append(rest)
```

```python
new_result
```

```python
import csv

with open('cleaned.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    for i in new_result:
        writer.writerow(i.split(","))
```
